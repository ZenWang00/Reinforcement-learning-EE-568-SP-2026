## Ex 1: Value Iteration [20 pts.]

### 1.1 Implement value iteration

````python
def value_iteration(env, tol=1e-10):
    """Inplementation of value iteration, note that the implementation is based on value iteration mentioned in the lecture.
    Args:
        env: environment
        tol: a scalar to dermerminate whether the algorithm convergences
    Returns:
        policies: policy  at each iteration (list)
        vs: value functions at each iteration (list)
    """
    policies = []
    vs = []
    v = np.zeros(env.n_states) # initialize value function
    q = np.zeros((env.n_states, env.n_actions)) #initialize Q-value
    
    while True:
        v_old = np.copy(v) # save a copy of value function for the convergence criterion at the step
        for a in range(env.n_actions):
            q[:, a] = env.r[:, a] + gamma * np.dot(env.T[a], v) # calculate Q-value
        v = np.max(q, axis=1) # update value function
        policies.append(np.argmax(q, axis=1)) # obtain policy
        vs.append(v)
        if np.linalg.norm(v - v_old) < tol: # convergence criterion
            break
    return policies, vs
````

````python
policies, vs = value_iteration(gridworld)
````

### 1.2: Evaluate the extracted policies

For the evaluation of the sequence of policies, i.e. computing $V^\pi$ for each policy $\pi$ we recall that $V^\pi$ is the fixed point $V^\pi = T^\pi (V^\pi)$.
In the following code, we ask you to apply the operator $T^\pi$ a number of time sufficient to find a value $V$ such that $ || T^\pi(V) - V || \leq \mathrm{tol}$

````python
def evaluate_policy_sequence(policies, env, tol=1e-10):
    """Inplementation of policy evaluation through iteratively applying policy value iteration 
    Args:
        policies: a list of policies obtained by section 1.1
        env: environment
        tol: a scalar to dermerminate whether the policy evaluation convergences
    Returns:
        values: a list of value function for each policy
    """
    values = []
    for pi in policies:
        v = np.zeros(env.n_states) # initialize value function
        q = np.zeros((env.n_states, env.n_actions)) #initialize Q-value
        while True:
            v_old = np.copy(v) # save a copy of value function for the convergence criterion at the step
            for a in range(env.n_actions):
                q[:, a] = env.r[:, a] + gamma * np.dot(env.T[a], v) #calculate Q-value
            for s in range(env.n_states):
                action_taken = pi[s] # obtain the action determined by the policy
                v[s] = q[s, action_taken] #calculate value function by $v(s) = Q(s,pi(s))$
            if np.linalg.norm(v - v_old) < tol: # convergence criterion
                break
        values.append(v)
    return values
````

````python
def compute_subopt(values, v_star):
    """
    Args:
        values: a list of value function.
        v_star: the optimal value function obtained by MDP.solver
    Returns:
        subopts: an array indicates the suboptimality.
    """
    subopts = []
    for v in values:
        i = np.argmax(np.abs(v - v_star))
        subopts.append(-v[i] + v_star[i])
    return np.array(subopts)
````

## 1.Proof that the Bellman optimality operator $T$ is a contraction

The Bellman optimality operator is defined as
$$
(TV)(s)=\max_{a\in\mathcal A}\left[r(s,a)+\gamma\sum_{s'}P(s'|s,a)V(s')\right].
$$

To prove that $T$ is a contraction in the sup norm, let $V,W$ be two arbitrary value functions. For any state $s$,
$$
\begin{aligned}
|(TV)(s)-(TW)(s)|
&=
\left|
\max_a \left[r(s,a)+\gamma\sum_{s'}P(s'|s,a)V(s')\right]
-
\max_a \left[r(s,a)+\gamma\sum_{s'}P(s'|s,a)W(s')\right]
\right|.
\end{aligned}
$$

Using the inequality
$$
|\max_a x_a-\max_a y_a|\le \max_a |x_a-y_a|,
$$
we obtain
$$
\begin{aligned}
|(TV)(s)-(TW)(s)|
&\le
\max_a
\left|
\gamma\sum_{s'}P(s'|s,a)\bigl(V(s')-W(s')\bigr)
\right| \\
&\le
\gamma \max_a \sum_{s'}P(s'|s,a)\,|V(s')-W(s')|.
\end{aligned}
$$

Now, by the definition of the sup norm,
$$
|V(s')-W(s')|\le \|V-W\|_\infty
\qquad \text{for all } s'.
$$
Therefore,
$$
\sum_{s'}P(s'|s,a)\,|V(s')-W(s')|
\le
\sum_{s'}P(s'|s,a)\,\|V-W\|_\infty
=
\|V-W\|_\infty,
$$
since $\sum_{s'}P(s'|s,a)=1$.

Hence,
$$
|(TV)(s)-(TW)(s)|\le \gamma \|V-W\|_\infty.
$$

Taking the maximum over all states $s$, we get
$$
\|TV-TW\|_\infty \le \gamma \|V-W\|_\infty.
$$

Since $0\le \gamma <1$, the Bellman optimality operator $T$ is a contraction mapping in the sup norm.

### 2.Explain value of iterations upper bound
The value iteration updates are defined by
$$
V_{k+1}=TV_k.
$$
To show that $V_k$ is an upper bound on $V^\star$, we use two facts.

First, the Bellman optimality operator $T$ is monotone: if $V\ge W$, then $TV\ge TW$.
Second, the optimal value function satisfies the Bellman equation
$$
V^\star = TV^\star.
$$

Therefore, if the initial value satisfies $V_0\ge V^\star$, then
$$
V_1=TV_0 \ge TV^\star = V^\star.
$$
Repeating the same argument inductively gives
$$
V_k \ge V^\star \quad \text{for all } k.
$$

This does not contradict
$$
V^\star=\max_{\pi\in\Pi}V^\pi,
$$
because $V^\star$ is the maximum over value functions induced by actual policies, whereas the intermediate iterates $V_k$ from value iteration are generally not the value function of any fixed policy. They are only intermediate optimistic estimates, so they may temporarily lie above $V^\star$ before converging to it.

# Ex 2: Policy Iteration [20 pts.]

````python
def evaluate_policy(pi, env, tol=1e-10):
    """Implementation of policy evaluation through iteratively applying using a certain policy 
    Args:
        pi: a policy
        env: environment
        tol: a scalar to dermerminate whether the policy evaluation convergences
    Returns:
        v: an array with the values of the actions chosen
        q: an array with the q values    
    """
    v = np.zeros(env.n_states)
    q = np.zeros((env.n_states, env.n_actions))
    while True:
        v_old = np.copy(v)
        for a in range(env.n_actions):
            q[:, a] = env.r[:, a] + gamma * np.dot(env.T[a], v_old) # calculate Q-value
        for s in range(env.n_states):
            v[s] = q[s, pi[s]] # calculate value function by $v(s) = Q(s,pi(s))$
        
        if np.linalg.norm(v - v_old) < tol:
            break
    return v, q
````

````python
def get_greedy_policy(q):
    """Implementation of a greedy approach to choose policies (policy improvement)
    Args:
        q: q values obtained from evaluating the policies
    Returns:
        policy: greedy policy (list)
    """
    policy = []
    for s in range(q.shape[0]):
        policy.append(np.argmax(q[s, :])) # choose the action with the largest q value
    return policy
````

````python
def policy_iteration(env, tol=1e-20):
    """Implementation of policy iteration
    Args:
        env: environment
        tol: a scalar to dermerminate whether the algorithm convergences
    Returns:
        vs: value functions at each iteration (list)
    """
    vs = []
    v = np.zeros(env.n_states)
    q = np.zeros((env.n_states, env.n_actions))
    pi = np.zeros(env.n_states, dtype=int)
    while True:
        v_old = np.copy(v)
        v, q = evaluate_policy(pi, env, tol)
        pi = get_greedy_policy(q)
        vs.append(v)
        if np.linalg.norm(v - v_old) < tol:
            break
    return vs, pi
````

# Ex 3: Modified Policy Iteration [20 pts.]

In this cell you will code and run Modified Value Iteration, that follows this pseudocode:
- $\pi_{k+1} = \mathcal{G}(V_k)$
- $V_{k+1} = T^m_{\pi_{k+1}}(V_k)$

In the implementation, we will use also the variable $Q$ as in the previous implementation of VI and PI for convenience of implementation.

**Questions**

Which value of $m$ recovers policy iteration ?


Which value of $m$ recovers value iteration ?



***Answers***

**Which value of $m$ recovers policy iteration ?**

When $m \to \infty$, it reduces to Policy Iteration.

According to the Lecture 1 slides page 41 and 43, because the Bellman expectation operator $\mathcal{T}^\pi$ is a $\gamma$-contraction mapping, applying this operator an infinite number of times will lead to convergence at its unique fixed point, $V^{\pi_{k+1}}$, which is the true value function under policy $\pi_{k+1}$. That is, $V_{k+1} = \mathcal{T}^m_{\pi_{k+1}}(V_k)$ becomes


$$V_{k+1} = \lim_{m \to \infty} (\mathcal{T}^{\pi_{k+1}})^m V_k = V^{\pi_{k+1}} \tag{1}$$


At this stage, $V_{k+1}$ becomes the exact true value function of the policy $\pi_{k+1}$. According to the Lecture 1 slides page 51, this is exactly the step 1 policy evaluation $V_t \leftarrow \mathcal{T}^{\pi_t} V_t$. In this step, it iteratively applies policy value iteration until convergence.

In the next step, it performs the policy improvement, which means updating the current policy $\pi_t$ by the greedy policy. According to the Lecture 1 slides page 51:
$$
\pi_{t+1}(s) = \arg\max_{a \in \mathcal{A}}
\left[
r(s,a) + \gamma \sum_{s' \in \mathcal{S}} P(s' \mid s,a)\, V^{\pi_t}(s')
\right]
\Longrightarrow
\pi_{k+1} = \mathcal{G}(V_k)
\tag{2}
$$

This is exactly $\pi_{k+1} = \mathcal{G}(V_k)$ in Modified Value Iteration.


Therefore, this is standard Policy Iteration, which consists of exact policy evaluation(equation(1)) and greedy improvement(equation(2)).

**Which value of $m$ recovers value iteration ?**

When $m = 1$, it reduces to Value Iteration.

According to the Lecture 1 sides page 46, the definition of value iteration is:

For each iteration $t$, apply the Bellman optimal operator $\mathcal{T}$ to the current value estimate $V_t$:

$$V_{t+1} = \mathcal{T}V_t$$

If $m=1$, we only apply the operator once to the newly updated policy:


$$V_{k+1} = \mathcal{T}^m_{\pi_{k+1}}(V_k) \Longrightarrow V_{k+1} = \mathcal{T}^{\pi_{k+1}} V_k \tag{1}$$




Based on greedy improvement $\pi_{k+1} = \mathcal{G}(V_k)$ in Modified Value Iteration, for each state $s$ it has (according to the Lecture 1 sides page 51):

$$\pi_{k+1}(s) \in \arg\max_{a \in \mathcal{A}} \left[ r(s,a) + \gamma \sum_{s'} P(s'|s,a) V_k(s') \right] \tag{2}$$


By substituting the definition of $\pi_{k+1}$ from the policy improvement (equation(2)) step into the equation(1), we get:


$$V_{k+1}(s)= (\mathcal{T}^{\pi_{k+1}} V_k)(s) = r(s, \pi_{k+1}(s)) + \gamma \sum_{s'} P(s' | s, \pi_{k+1}(s)) V_k(s') \tag{3}$$


Since $\pi_{k+1}(s)$ is the specific action $a$ that maximizes the inner term (which means it is the $\arg\max$), the equation(3) above is mathematically equivalent to:


$$V_{k+1}(s) = (\mathcal{T}^{\pi_{k+1}} V_k)(s) = \max_{a \in \mathcal{A}} \left[ r(s, a) + \gamma \sum_{s'} P(s' | s, a) V_k(s') \right] \tag{4}$$

By observing the definition of Bellman optimality operator (according to the Lecture 1 sides page 43),

$$(\mathcal{T} V)(s) = \max_{a \in \mathcal{A}} \left[ r(s, a) + \gamma \sum_{s' \in \mathcal{S}} P(s' | s, a) V(s') \right] \tag{5}$$

We discover that the right hand side of equation(4) is exactly the definition of the Bellman optimality operator, meaning $V_{k+1} = \mathcal{T}^{\pi_{k+1}} V_k = \mathcal{T} V_k$. Therefore, this algorithm becomes standard Value Iteration.

First, write a function that applies the Bellman evaluation operator $m$ times

````python
def policy_operator_m_times(pi, m, env, v, q):
    """Implementation of partial policy evaluation through applying m times the Bellman operator
    Args:
        pi: a policy
        env: environment
        tol: a scalar to dermerminate whether the policy evaluation convergences
        v: initial value vector
        q: initial state action value vector
    Returns:
        v: an array with the values of the actions chosen
        q: an array with the q values    
    """
    i = 0
    while i < m:
        # Bellman expectation update for Q under the current value v
        for a in range(env.n_actions):
            q[:, a] = q[:, a] = env.r[:, a] + env.gamma * np.dot(env.T[a], v)
        
        # Policy evaluation step: V(s) = Q(s, pi(s))
        for s in range(env.n_states):
            v[s] = q[s, pi[s]]
        i = i + 1
    return v, q
````

At this point, using the function `get_greedy_policy` implemented for PI you are ready to implement Modified Policy Iteration !

````python
def modified_policy_iteration(env, m, tol=1e-10):
    """Implementation of modified policy iteration
    Args:
        env: environment
        tol: a scalar to dermerminate whether the algorithm convergences
    Returns:
        vs: value functions at each iteration (list)
    """
    vs = []
    policies = []
    v = np.zeros(env.n_states)
    q = np.zeros((env.n_states, env.n_actions))
    pi = np.zeros(env.n_states, dtype=int)
    while True:
        v_old = np.copy(v)
        # for a in range(env.n_actions):
        #     q[:, a] = env.r[:, a] + env.gamma * np.dot(env.T[a], v)
        pi = get_greedy_policy(q) # Greedy improvement step, should it use the q value obtained from the previous policy evaluation step?
        v, q = policy_operator_m_times(pi, m, env, v, q) # Partial policy evaluation step
        policies.append(np.copy(pi))
        vs.append(v)
        if np.linalg.norm(v - v_old) < tol:
            break
    return vs,policies
````

**Questions** 
- How does the value of $m$ affect convergence ?
- Which is the cost per iteration of Modified Policy Iteration ?
- Which is the cost per iteration of Policy Iteration implemented with Option 2 in the slides ?

***Answers***

**How does the value of $m$ affect convergence ?**

For convergence, increasing $m$ usually makes the number of modified policy iterations (`while` loop of function `modified_policy_iteration`) smaller, but the number of the inner iterations (`while` loop inside `policy_operator_m_times` function) larger.

When $m=1$, modified policy iteration is just value iteration, so it needs the most modified policy iterations. 
As $m$ grows, each policy is evaluated (`policy_operator_m_times`) more accurately before the next greedy improvement, so the method behaves more and more like policy iteration. With limit of `m` to infinity, it becomes “full” policy evaluation which means that modified policy iteration becomes policy iteration.

The plot matches our explanation: $m=1$ converges more slowly in terms of modified policy iterations, while larger $m$ drops the suboptimality much faster. But the tradeoff is that modified policy iteration becomes more expensive, so a larger $m$ is not automatically best overall.

**Which is the cost per iteration of Modified Policy Iteration ?**
The cost per iteration of Modified Policy Iteration can be divided into two parts: the partial policy evaluation step `policy_operator_m_times`, and the greedy policy improvement step `get_greedy_policy`.

First, consider `policy_operator_m_times`. Inside this function, the main computation is

```python
for a in range(env.n_actions):
    q[:, a] = env.r[:, a] + env.gamma * np.dot(env.T[a], v)
```

For one fixed action $a$, the matrix `env.T[a]` has size $|S| \times |S|$, and the vector $v$ has size $|S|$. Therefore, `np.dot(env.T[a], v)` is a matrix vector multiplication. For a dense matrix, this operation costs $O(|S|^2)$. Since this is repeated for all $|A|$ actions, the total cost of this part is

$$
O(|S|^2 |A|).
$$

Then the function updates the value function by selecting the Q value of the action chosen by the policy:

```python
for s in range(env.n_states):
    v[s] = q[s, pi[s]]
```

This loop goes through all states once, and each operation inside the loop is constant time. So this part costs

$$
O(|S|).
$$

Hence, one application of the operator inside `policy_operator_m_times` costs

$$
O(|S|^2 |A|) + O(|S|).
$$

Since the function applies this operator $m$ times, the total cost of `policy_operator_m_times` is

$$
O\big(m(|S|^2 |A| + |S|)\big).
$$

Because the term $O(|S|)$ is much smaller than $O(|S|^2 |A|)$, it can be absorbed into the dominant term. So we usually simplify this to

$$
O(m|S|^2 |A|).
$$

Now consider the greedy policy improvement step `get_greedy_policy(q)`:

```python
for s in range(q.shape[0]):
    policy.append(np.argmax(q[s, :]))
```

For each state $s$, the function computes `np.argmax(q[s, :])`, which scans all actions in that row. This costs $O(|A|)$ for one state. Since the loop runs over all $|S|$ states, the total cost is

$$
O(|S||A|).
$$

Therefore, the cost of one outer iteration of Modified Policy Iteration is the sum of these two parts:

$$
O(m|S|^2 |A|) + O(|S||A|).
$$

So the final complexity per iteration is

$$
O(m|S|^2 |A| + |S||A|).
$$

If we only keep the dominant term, then the cost per iteration can be written as

$$
O(m|S|^2 |A|).
$$

All in all, the partial policy evaluation step is the most expensive part, while the greedy improvement step is smaller. 

We can say that the complexity of modified policy iteration is mainly determined by how many times we apply the Bellman operator, that is, by the value of $m$.


**Which is the cost per iteration of Policy Iteration implemented with Option 2 in the slides ?**
For policy iteration with option 2 in the slides, the policy evaluation step uses the closed-form solution

$$
V^\pi=(I-\gamma P^\pi)^{-1}R^\pi,
$$

and the lecture slides give its computation cost as

$$
O(|S|^3 + |S|^2|A|).
$$

This is because the $O(|S|^3)$ part comes from solving or inverting the $|S|\times |S|$ linear system, and the $O(|S|^2|A|)$ part is from building the policy-dependent quantities and doing the greedy improvement. This can be divided into two parts: policy evaluation by the closed-form solution and policy improvement by the greedy update.

For a fixed policy $\pi$, the lecture slides give the matrix form of the Bellman consistency equation as

$$
V^\pi = R^\pi + \gamma P^\pi V^\pi.
$$

Rearranging this equation gives

$$
(I-\gamma P^\pi)V^\pi = R^\pi,
$$

and therefore

$$
V^\pi = (I-\gamma P^\pi)^{-1}R^\pi.
$$

This is exactly option 2 in the slides. To compute $V^\pi$, we must solve a linear system with matrix $I-\gamma P^\pi$, whose size is $|S|\times |S|$. 

For a dense matrix, solving or inverting such a system costs

$$
O(|S|^3).
$$

This is the first main part of the complexity.

Then, we explain where the second term comes from. 

Before solving the system, we need to construct the policy-dependent reward vector $R^\pi$ and transition matrix $P^\pi$. Their definitions are

$$
R^\pi_s := \sum_{a\in A}\pi(a|s)r(s,a),
$$

and

$$
P^\pi_{s,s'} := \sum_{a\in A}\pi(a|s)P(s'|s,a).
$$

To build $R^\pi$, for each state $s$ we sum over all actions, so this costs about

$$
O(|S||A|).
$$

To build $P^\pi$, for each pair $(s,s')$ we sum over all actions. Since there are $|S|^2$ pairs of states and $|A|$ actions, this costs

$$
O(|S|^2|A|).
$$

Since $O(|S|^2|A|)$ dominates $O(|S||A|)$, so the simplified construction cost is

$$
O(|S|^2|A|).
$$

After policy evaluation, policy iteration still needs the policy improvement step. The greedy update is

$$
\pi_{t+1}(s)=\arg\max_{a\in A}
\left[
r(s,a)+\gamma\sum_{s'\in S}P(s'|s,a)V^{\pi_t}(s')
\right].
$$

To compute this for one state-action pair $(s,a)$, we must sum over all next states $s'$, which costs

$$
O(|S|).
$$

Since this must be done for all $|S|$ states and all $|A|$ actions, the total cost of greedy improvement is

$$
O(|S|^2|A|).
$$

Finally, adding the first part and second part together, the total per-iteration cost of one iteration of policy iteration with option 2 is

$$
O(|S|^3 + |S|^2|A|).
$$

In conclusion, policy iteration with option 2 is usually more expensive per iteration than value iteration or modified policy iteration, especially when the number of states is large.

### Q learning with epsilon greedy 

In this subsection, we implement $Q$ Learning with the simplest possible exploration strategy: $\epsilon$-greedy.

````python
def argmax_with_random_tie_breaking(b):
    return np.random.choice(np.where(b == b.max())[0])
````

````python
def q_learning_epsilon_greedy(K: int = 100000, H: int = 20, epsilon: float = 0.1, alpha: float = 0.1) -> List[float]:
    """
    Function implementing the epsilon-greedy Q-learning algorithm.

    :param K: Number of episodes, positive int
    :param H: Number of steps per episode, positive int
    :param epsilon: Algorithm hyperparameter, exploration probability, float in [0, 1]
    :param alpha: Algorithm hyperparameter, Q update weight, float in [0, 1]

    :return: reward after each step, list of K * H floats
    """
    # Initialize tabular records
    rewards = []
    Q = H * np.ones((H, gridworld.n_states, gridworld.n_actions))
    V = H * np.ones((H + 1, gridworld.n_states))
    V[H, :] = 0

    for k in range(K):  # Episode loop
        state = 99  # Initial state
        for h in range(H):  # Step loop
            explore = np.random.binomial(2, p=epsilon)
            if explore:
                # Exploration: With probability epsilon take a random action
                a = np.random.choice(gridworld.n_actions)
            else:
                # Exploitation: With probability 1 - epsilon take one of the optimal actions for the current state
                a = argmax_with_random_tie_breaking(Q[h, state, :])

            # Get reward for action
            rewards.append(gridworld.r[state, a])

            # Get the new state according to the transition dynamics
            new_state = np.random.choice(gridworld.n_states,
                                         p=gridworld.T[a][state])

            # Update Q according to the algorithm
            Q[h, state, a] = (1 - alpha) * Q[h, state, a] + alpha * (gridworld.r[state, a] + V[h + 1, new_state]
)


            # Update V as the Q-value of the optimal actions for the current state
            V[h, state] = np.min([np.max(Q[h, state, :]), H])

            state = new_state
    return rewards
````

### Q-Learning with bonuses

In this subsection, you are asked to implement an optimistic version of  Q Learning suggested in (Jin et al., 2018)

````python
def q_learning_ucb_bonuses(K: int = 100000, H: int = 20, c: float = 0.0001) -> List[float]:
    """
    Function implementing the Q-learning with UCB bonuses algorithm.

    :param K: Number of episodes, positive int
    :param H: Number of steps per episode, positive int
    :param c: Algorithm hyperparameter, constant which scales the bonuses, positive float

    :return: reward after each step, list of K * H floats
    """

    # Initialize tabular records
    rewards = []
    Q = H * np.ones((H, gridworld.n_states, gridworld.n_actions))
    V = H * np.ones((H + 1, gridworld.n_states))
    V[H, :] = 0
    N = np.ones((H, gridworld.n_states, gridworld.n_actions))

    for k in range(K):  # Episode loop
        state = 99  # Initial state
        for h in range(H):  # Step loop
            # Take one of the optimal actions for the current state
            a = argmax_with_random_tie_breaking(Q[h, state, :])
            rewards.append(gridworld.r[state, a])

            # Record that we visited this state-action pair (again)
            N[h, state, a] += 1

            # Get the new state according to the transition dynamics
            new_state = np.random.choice(gridworld.n_states,
                                         p=gridworld.T[a][state])

            # Calculate the UCB bonus
            bonus = c * np.sqrt(H ** 3 / N[h, state, a])

            # Calculate the adaptive alpha according to the algorithm
            alpha = (H + 1) / (H + N[h, state, a])

            # Update Q according to the algorithm
            Q[h, state, a] = (1 - alpha) * Q[h, state, a] + alpha * (gridworld.r[state, a] + V[h + 1, new_state] + bonus)

            # Update V as the Q-value of the optimal actions for the current state
            V[h, state] = np.max(Q[h, state, :])
            V[h, state] = np.clip(V[h, state], 0, H)

            state = new_state

    return rewards
````

**Question:** You are given an environment with $N$ states linked in a chain, i.e. from every state except the two endpoints there are two possible actions going right or go back to the first state in the chain. For the endpoints the second action is to stay in the same state. The reward is only 1 if you are in the rightmost state and stay there, otherwise 0. What is the probability that the $\varepsilon$-greedy algorithm will have a positive reward after $H$ steps in the first episode, if the chosen initial state is the leftmost state in the chain? For simplicity you can consider the case $H=N$.

The environment consists of $N$ states in a chain, $\{s_1, s_2, \dots, s_N\}$, with a starting state of $s_1$ and a horizon $H=N$. At any state except the endpoints, the agent can either move right to the next state or go back to the start. The only way to receive a positive reward of **1** is to reach the final state $s_N$ and choose the action to stay there. 

Since the horizon $H$ is exactly equal to the number of states $N$, there is only one specific sequence of actions that results in success: the agent must move right $N-1$ times to reach $s_N$, and on the $N$-th step, perform the action to stay at $s_N$. Any deviation will exhaust the step limit and result in a reward of 0.

In the first episode, we assume all Q-values are initialized equally (typically to zero). Under the $\varepsilon$-greedy strategy, the agent exploits with probability $1-\varepsilon$ and explores with probability $\varepsilon$. Because all actions have tied values initially, both exploitation (breaking ties randomly) and exploration will result in picking an action uniformly at random. 

Therefore, the probability of picking the correct action at any given state is exactly 1/2. To get a positive reward, the agent must make this correct choice $N$ times in a row. The probability is calculated as:

$$P(\text{Success})=\prod_{t=1}^{N}P(\text{correct action at step }t)$$

$$P(\text{Success})=\left(\frac{1}{2}\right)^N$$

This highlights the curse of dimensionality in sparse reward environments: as the state space $N$ grows, the probability of finding the reward through naive exploration decreases exponentially.

````python
def generate_episode(policy: List, H: int, initial_state: int):
    """
    Generate an episode using the given policy.

    :param policy: A list where each element is a probability distribution over actions for a state.
    :param H: Maximum number of steps in the episode.
    :param initial_state: Starting state.

    :return: A list of (state, action, reward) tuples.
    """
    episode = []
    state = initial_state
    for _ in range(H):
        action = policy[state]
        reward = gridworld.r[state, action]
        next_state = np.random.choice(gridworld.n_states, p=gridworld.T[action][state])
        episode.append((state, action, reward))
        state = next_state
    return episode

def mc_policy_evaluation_with_linear_approx(
    policy: List[np.ndarray], 
    phi: np.ndarray,
    K: int = 1000, 
    H: int = 20, 
    gamma: float = 0.99, 
    alpha: float = 0.01
) -> np.ndarray:
    """
    Monte Carlo policy evaluation with linear function approximation.

    :param policy: A list where each element is a probability distribution over actions for a state.
    :param K: Number of episodes.
    :param H: Maximum number of steps per episode.
    :param gamma: Discount factor.
    :param alpha: Learning rate for gradient descent.

    :return: The learned weight vector theta.
    """
    # Feature matrix: Each state is represented by a one-hot vector
    n_states = gridworld.n_states

    # Initialize weight vector theta
    theta = np.zeros(n_states)

    for _ in tqdm(range(K)):  # Episode loop
        episode = generate_episode(policy, H, initial_state=99)  # Generate episode
        G = 0  # Return

        # Backward pass through the episode
        for t in range(len(episode) - 1, -1, -1):
            state, _, reward = episode[t]
            G = gamma * G + reward  # Discounted return

            # Update theta using gradient descent
            phi_s = phi[state]  # Feature vector for the current state
            V_hat = np.dot(theta, phi_s)  # Current value estimate
            theta += alpha * (G - V_hat) * phi_s  # Gradient descent update

    return theta
````

````python

import numpy as np
def td_policy_evaluation_with_linear_approx(
    policy: List[np.ndarray], 
    phi: np.ndarray,
    K: int = 1000, 
    H: int = 20, 
    gamma: float = 0.99, 
    alpha: float = 0.01
) -> np.ndarray:
    """
    Temporal Difference (TD) policy evaluation with linear function approximation.

    :param policy: A list where each element is a probability distribution over actions for a state.
    :param K: Number of episodes.
    :param H: Maximum number of steps per episode.
    :param gamma: Discount factor.
    :param alpha: Learning rate for gradient descent.

    :return: The learned weight vector theta.
    """
    # Feature matrix: Each state is represented by a one-hot vector
    n_states = gridworld.n_states

    # Initialize weight vector theta
    theta = np.zeros(n_states)

    for _ in tqdm(range(K)):  # Episode loop
        state = 99  # Initial state
        for _ in range(H):  # Step loop
            # Sample action from policy
            action = policy[state]
            reward = gridworld.r[state, action]
            next_state = np.random.choice(gridworld.n_states, p=gridworld.T[action][state])

            # Compute TD target and TD error
            phi_s = phi[state]  # Feature vector for current state
            phi_s_next = phi[next_state]  # Feature vector for next state
            V_s = np.dot(theta, phi_s)  # Current value estimate
            V_s_next = np.dot(theta, phi_s_next)  # Next value estimate
            td_target = reward + gamma * V_s_next  # TD target
            td_error = td_target - V_s  # TD error

            # Update theta using gradient descent
            theta += alpha * td_error * phi_s
            # Move to the next state
            state = next_state

    return theta
````
